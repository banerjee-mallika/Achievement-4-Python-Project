```python
#The mixed column if filled with different types of data 
df_test['mix'] = ['a', 'b', 1, True]
```

```python
#Creating fuction to search for a mixed data type in a column
for col in df_test.columns.tolist():
  weird = (df_test[[col]].applymap(type) != df_test[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_test[weird]) > 0:
    print (col)
```

## Observation: 
After conducting exploratory analysis on the df_products DataFrame, outliers were detected in the prices column. The column had a mean of 9.99 and a high standard deviation of 453.6, which is well beyond two standard deviations from the mean—indicating the presence of extreme values.

Specifically, two outlier values were identified: 14,900 and 99,999. These values were more than two standard deviations above the mean. To address this, they were replaced with the median price of 7.1.

Given the wide range between the minimum value 1 and the maximum value 99999, using the median is more appropriate than the mean for handling outliers, as it is less influenced by extreme values.

After replacing the outliers with the median, the overall mean of the prices column dropped to 7.6, reflecting a more accurate representation of the dataset.
