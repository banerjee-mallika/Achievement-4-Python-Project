# Merging Dataframe

Observation: The column 'product_id' is a common key in both DataFrames, indicating that a merge can be performed. The appropriate type of merge depends on the desired outcome. Since we are only interested in rows where 'product_id' exists in both DataFrames, an inner merge is the most suitable choice. This ensures that only matching records are retained, making the resulting data more meaningful and relevant.


```python
# test merge without overwriting with merge flag 'match'

pd.merge(orders_products_combined , df_prods , on = ['product_id'] , how = 'inner', indicator = 'match' )
```

Note: In the previous merge operation, a 'merge' column was created as a flag to indicate the source of each row. However, this column is not useful for the new merge between orders_products_combined and df_prods. For better presentation and clarity, the 'merge' column from the earlier operation should be dropped after completing the new merge.

```python
# actual merge of 'orders_products_combined' and '_df_pords'
ords_prods_merge = orders_products_combined.merge(df_prods , on = ['product_id'] , indicator = 'match' )
```

```python
# drop the flag column '_merge' from the previous merge
ords_prods_merge.drop('_merge', axis = 1 )
```
